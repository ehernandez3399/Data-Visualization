## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
import matplotlib.pyplot as plt

QUERY = """
SELECT ConvertedComp
FROM master
"""
pd.read_sql_query(QUERY, conn)

df = pd.read_sql_query(QUERY,conn)

plt.figure(figsize=(10, 6))
plt.hist(df['ConvertedComp'], bins=30, edgecolor='black')  # Adjust bins as needed
plt.xlabel('Converted Compensation')
plt.ylabel('Frequency')
plt.title('Histogram of Converted Compensation')
plt.grid(True)
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
import seaborn as sns

QUERY = """
SELECT Age
FROM master
"""
pd.read_sql_query(QUERY, conn)

df = pd.read_sql_query(QUERY,conn)

plt.figure(figsize=(8, 6))
sns.boxplot(x='Age', data=df)
plt.title('Boxplot of Age')
plt.xlabel('Age')
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
QUERY = """
SELECT Age, WorkWeekHrs
FROM master
"""
pd.read_sql_query(QUERY, conn)

df = pd.read_sql_query(QUERY,conn)

plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['WorkWeekHrs'], alpha=0.5)
plt.title('Scatter Plot of Age vs WorkWeekHrs')
plt.xlabel('Age')
plt.ylabel('WorkWeekHrs')
plt.grid(True)
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM master
"""
pd.read_sql_query(QUERY, conn)

df = pd.read_sql_query(QUERY,conn)

x = df['WorkWeekHrs']
y = df['CodeRevHrs']
bubble_size = df['Age'] * 10 

plt.figure(figsize=(10, 8))
plt.scatter(x, y, s=bubble_size, alpha=0.5, cmap='viridis')
plt.title('Bubble Plot of WorkWeekHrs vs CodeRevHrs')
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.colorbar(label='Age')
plt.grid(True)
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
QUERY = """
SELECT *
FROM DatabaseDesireNextYear
Order BY Respondent DESC LIMIT 5
"""
pd.read_sql_query(QUERY, conn)

df = pd.read_sql_query(QUERY,conn)

df

database_names = df['DatabaseDesireNextYear']
percentages = df['Respondent'] / df['Respondent'].sum() * 100 

plt.figure(figsize=(8, 8))
plt.pie(percentages, labels=database_names, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
plt.axis('equal')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs
FROM master
WHERE Age BETWEEN 30 AND 35
"""
pd.read_sql_query(QUERY, conn)

df = pd.read_sql_query(QUERY,conn)

median_values = df.median()

# Plotting the stacked bar chart using Matplotlib
plt.figure(figsize=(10, 6))

# Plotting WorkWeekHrs
plt.bar('Median Values', median_values['WorkWeekHrs'], label='WorkWeekHrs', color='b')

# Plotting CodeRevHrs on top of WorkWeekHrs to create a stacked chart
plt.bar('Median Values', median_values['CodeRevHrs'], bottom=median_values['WorkWeekHrs'], label='CodeRevHrs', color='g')

plt.title('Stacked Chart of Median WorkWeekHrs and CodeRevHrs for Ages 30-35')
plt.legend()
plt.grid(True)
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
QUERY = """
SELECT Age, ConvertedComp
FROM master
WHERE Age BETWEEN 45 AND 60
"""
pd.read_sql_query(QUERY, conn)

df = pd.read_sql_query(QUERY,conn)

median_values = df.groupby('Age')['ConvertedComp'].median()

# Plotting the median ConvertedComp for ages 45 to 60
plt.figure(figsize=(10, 6))
plt.plot(median_values.index, median_values.values, marker='o', linestyle='-', color='b')
plt.xlabel('Age')
plt.ylabel('Median Converted Compensation')
plt.title('Median Converted Compensation for Ages 45 to 60')
plt.grid(True)
plt.xticks(median_values.index)  # Set x-axis ticks to match age values
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
QUERY = """
SELECT MainBranch
FROM master
"""
pd.read_sql_query(QUERY, conn)

df = pd.read_sql_query(QUERY,conn)

branch_counts = df['MainBranch'].value_counts()

# Plotting the horizontal bar chart using Matplotlib
plt.figure(figsize=(10, 6))
plt.barh(branch_counts.index, branch_counts.values, color='skyblue')
plt.xlabel('Number of Respondents')
plt.ylabel('MainBranch')
plt.title('Distribution of MainBranch')
plt.grid(True)
plt.show()


```
